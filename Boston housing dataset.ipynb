```python
ones = np.ones((x.shape[0],1))
x = np.hstack((ones, x))
print(x.shape)

```

```python
def hypothesis(x, theta):
    y_ = 0.0
    n = x.shape[0]
    for i in range(n):
        y_ += theta[i]*x[i]
    return y_

def error(x, y, theta):
    total_error = 0.0
    m = x.shape[0]
    for i in range(m):
        y_ = hypothesis(x[i], theta)
        total_error += (y_ - y[i])**2
    return total_error/m

def gradient(x, y , theta):
    m,n = x.shape
    grad = np.zeros((n,))
    for j in range(n):
        for i in range(m):
            y_ = hypothesis(x[i], theta)
            grad[j] += (y_-y[i])*x[i][j] 
    return grad/m
def gradient_descent(x, y, learning_rate = 0.1, maxSteps = 300):
    m,n = x.shape
    theta = np.zeros((n,))
    error_list = []
    for i in range(maxSteps):
        e = error(x, y, theta)
        error_list.append(e)
        grad = gradient(x, y, theta)
        for j in range(n):
            theta[j] = theta[j] - learning_rate*grad[j]
    return theta, error_list
```

```python
theta, error_list = gradient_descent(x, y)
```
